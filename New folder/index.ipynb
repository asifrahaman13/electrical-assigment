```python
# Impoprt all the necessary libraries and modules in the code
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read,write
from IPython.display import Audio
from numpy.fft import fft,ifft,irfft
import matplotlib.pyplot as plt
from scipy.io import wavfile as wav
import scipy.io.wavfile
import scipy.signal
import matplotlib as mpl
```

```python
# read data from the WAV file
rate, data = scipy.io.wavfile.read('../sound/second.wav') #preamble.wav r the a.wav file to be imported from the file
```

```python
# Play the file
Audio(data,rate=rate)
```

```python
print(f"Length of the data: {len(data)}")
```

```python
print(f"rate of the frequency: {rate}")
```

```python
times = np.arange(len(data))/rate
```

```python
print(f"times {times}")
```

```python
#Plot the time vs aplitude graph of original plot
plt.figure(figsize=(50, 20))
plt.plot(times, data)
plt.title("Original Signal with Noise",fontsize=50)
# Providing margins to the plot
plt.margins(0.01,0.1)
plt.show()
```

```python
plt.figure(figsize=(50,20))
print(f"original amplitudes: {data}")
times = np.arange(len(data))/rate
# Applying the fouries transoformation
fft_out = fft(data)
plt.title("Signal in frequency domain",fontsize=50)
# Plotting the signal in frequency domain
plt.plot(times, np.abs(fft_out))
# Providing margins to the plot
plt.margins(0.01,0.1)
plt.show()
```

```python
# apply a 3-pole lowpass filter at 0.1x Nyquist frequency
b, a = scipy.signal.butter(1, 0.1)
# Filtereing the plot
filtered = scipy.signal.filtfilt(b, a, data)
```

```python
plt.figure(figsize=(50,20))
print(data)
fft_out = fft(filtered)
# Plotting the signal in frequenccy domain after the removal of the noise
plt.plot(times, np.abs(fft_out))
plt.title("Signal in frequency domain after removal of noise",fontsize=50)
plt.margins(0.01,0.1)
plt.show()
```

```python
plt.figure(figsize=(50, 20))
plt.plot(times, data)
plt.title("Original signal with Noise",fontsize=50)
plt.margins(0.01,0.1)
plt.show()


plt.figure(figsize=(50,20))
new_sig = irfft(fft_out)
plt.plot(new_sig)
plt.title("Filetered signnal",fontsize=50)
plt.margins(0.01,0.1)
plt.show()


mpl.rc('xtick', labelsize=2) 
mpl.rc('ytick', labelsize=2)
```
